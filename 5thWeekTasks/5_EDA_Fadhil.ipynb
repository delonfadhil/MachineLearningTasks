Untuk melakukan Exploratory Data Analysis (EDA) pada dataframe di Python, Anda dapat mengikuti langkah-langkah berikut:

1. **Memahami Struktur Data**

  * Gunakan visualisasi seperti box plot untuk mengidentifikasi outliers.

  * Gunakan duplicated() untuk menemukan baris-baris yang duplikat.
  * Gunakan drop_duplicates() untuk menghapus duplikat jika diperlukan.

5. **Membuat Visualisasi Data**
  * Gunakan visualisasi seperti histogram, box plot, scatter plot, dan lainnya untuk memahami distribusi dan hubungan antar variabel.
  * Gunakan matplotlib atau seaborn untuk membuat visualisasi.

6. **Menganalisis Korelasi**
  * Gunakan corr() untuk menghitung korelasi antar kolom.
  * Gunakan heatmap untuk memvisualisasikan korelasi antar kolom.

7. **Menjalankan Uji Hipotesis (jika diperlukan)**
  * Gunakan uji statistik seperti uji t atau uji ANOVA untuk menguji hipotesis terkait data.

8. **Membuat Ringkasan dan Kesimpulan**
  * Buat ringkasan dari temuan utama Anda.
  * Buat visualisasi atau grafik ringkas untuk menyoroti temuan utama.

Prompt ChatGPT : beri kode untuk menampilkan pie diagram chart untuk fitur Age dengan min 20, max 69

```python
import matplotlib.pyplot as plt

# Pengelompokan data berdasarkan kategori usia
bins = [20, 30, 40, 50, 60, 69]
labels = ['20-30', '31-40', '41-50', '51-60', '61+']
df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

# Menghitung jumlah data dalam setiap grup
age_group_counts = df['Age_Group'].value_counts()

# Membuat pie chart
plt.figure(figsize=(8, 8))
plt.pie(age_group_counts, labels=age_group_counts.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Memastikan pie chart berbentuk lingkaran
plt.title('Distribution of Age Groups')
plt.show()
```

Prompt ChatGPT : buatlah uji statistik seperti uji t atau uji ANOVA untuk menguji hipotesis terkait data

```python
# Uji T antara dua kelompok:
from scipy.stats import ttest_ind

group1 = df[df['Main programming languages'] == 'Python']['Yearly brutto salary']
group2 = df[df['Main programming languages'] == 'Java']['Yearly brutto salary']

t_stat, p_value = ttest_ind(group1, group2)

print(f"T-statistic: {t_stat}, p-value: {p_value}")
if p_value < 0.05:
    print("Ada perbedaan signifikan antara kedua kelompok.")
else:
    print("Tidak ada perbedaan signifikan antara kedua kelompok.")
```

```python
# Uji ANOVA untuk membandingkan rata-rata antara beberapa kelompok:
from scipy.stats import f_oneway

groups = df.groupby('Main programming languages')['Yearly brutto salary']
f_stat, p_value = f_oneway(*[group for name, group in groups])

print(f"F-statistic: {f_stat}, p-value: {p_value}")
if p_value < 0.05:
    print("Terdapat perbedaan signifikan antara setidaknya dua kelompok.")
else:
    print("Tidak ada perbedaan signifikan antara kelompok-kelompok tersebut.")
```
