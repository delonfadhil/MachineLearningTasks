## Prompt ChatGPT : berikan kode untuk membuat model decision tree dengan dataset Iris.csv

```python
# Import Library
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import plot_tree
```

```python
# Load dataset Iris.csv dari google drive
iris_data = pd.read_csv('/content/drive/MyDrive/Dataset/iris.csv')
iris_data.head()
```

```python
# Pisahkan fitur (features) dan target (target) dari dataset
X = iris_data.drop('species', axis=1)
y = iris_data['species']
```

```python
# Bagi dataset menjadi data latih dan data uji
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Buat decision tree classifier
clf = DecisionTreeClassifier()
```

```python
# Latih lassifier dengan training set
clf.fit(X_train, y_train)
```

```python
# Buat prediksi dengan test set
y_pred = clf.predict(X_test)
```

```python
# Evaluasi classifier, classification report, dan confusion matrix
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

## Prompt ChatGPT : berikan kode untuk menvisualisasikannya dalam bentuk decision tree dan confusion matrix

```python
# Visualisasi dengan decision tree
plt.figure(figsize=(12, 8))
plot_tree(clf, feature_names=X.columns, class_names=y.unique(), filled=True)
plt.show()

# Visualisasi dengan confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, cmap='Blues', fmt='g', xticklabels=y.unique(), yticklabels=y.unique())
plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.title('Confusion Matrix')
plt.show()
```
