```python
# Menampilkan boxplot untuk fitur numerik
numerical_features = ['PassengerId', 'Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']
for feature in numerical_features:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x='Survived', y=feature, data=df)
    plt.title(f'Boxplot of {feature}')
    plt.show()
```

# Violin Plot

Prompt ChatGPT : beri kode untuk menampilkan violin plot dari fitur-fitur numerik

```python
# Menampilkan violin plot untuk fitur numerik
numerical_features = ['PassengerId', 'Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']
for feature in numerical_features:
    plt.figure(figsize=(8, 6))
    sns.violinplot(x='Survived', y=feature, data=df)
    plt.title(f'Violin Plot of {feature}')
    plt.show()
```

# Distributions

Prompt ChatGPT : beri kode untuk menampilkan visualisasi Distributions

```python
# Menampilkan distribusi untuk fitur numerik
numerical_features = ['PassengerId', 'Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']
for feature in numerical_features:
    plt.figure(figsize=(8, 6))
    sns.histplot(df[feature].dropna(), kde=True)
    plt.title(f'Distribution of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.show()
```

# Scatter Plot

Prompt ChatGPT : beri kode untuk menampilkan visualisasi scatter plot

```python
# Menampilkan scatter plot untuk fitur numerik
numerical_features = ['PassengerId', 'Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']
for i in range(len(numerical_features)):
    for j in range(i+1, len(numerical_features)):
        plt.figure(figsize=(8, 6))
        sns.scatterplot(x=numerical_features[i], y=numerical_features[j], data=df)
        plt.title(f'Scatter Plot of {numerical_features[i]} vs {numerical_features[j]}')
        plt.xlabel(numerical_features[i])
        plt.ylabel(numerical_features[j])
        plt.show()
```

# Line Plot

Prompt ChatGPT : beri kode untuk menampilkan visualisasi line plot

```python
# Misalnya, kita ingin membuat line plot untuk menampilkan perubahan rata-rata harga tiket (Fare) dari setiap kelas (Pclass) dari waktu ke waktu
fare_by_class = df.groupby('Pclass')['Fare'].mean().reset_index()

plt.figure(figsize=(8, 6))
sns.lineplot(x='Pclass', y='Fare', data=fare_by_class, marker='o')
plt.title('Average Fare by Class')
plt.xlabel('Class')
plt.ylabel('Average Fare')
plt.show()
```

# Bar Plot

Prompt ChatGPT : beri kode untuk menampilkan visualisasi bar plot

```python
# Misalnya, kita ingin membuat bar plot untuk menampilkan jumlah penumpang yang selamat (Survived=1) dan tidak selamat (Survived=0) berdasarkan jenis kelamin (Sex)
survived_by_sex = df.groupby('Sex')['Survived'].sum().reset_index()

plt.figure(figsize=(8, 6))
sns.barplot(x='Sex', y='Survived', data=survived_by_sex)
plt.title('Number of Survivors by Sex')
plt.xlabel('Sex')
plt.ylabel('Number of Survivors')
plt.show()
```

# Heat Map

Prompt ChatGPT : beri kode untuk menampilkan visualisasi heat map

```python
# Menampilkan heatmap dari korelasi antara fitur numerik
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```
