```python
# Memilih fitur-fitur numerik
numeric_features = ['carat', 'depth', 'table', 'price', 'x', 'y', 'z']

# Membuat boxplot untuk setiap fitur numerik
plt.figure(figsize=(12, 8))
for i, feature in enumerate(numeric_features, 1):
    plt.subplot(3, 3, i)
    sns.boxplot(x=df[feature])
    plt.xlabel(feature)

plt.tight_layout()
plt.show()
```

# Scatter Plot

Prompt ChatGPT : beri kode untuk menampilkan scatter plot untuk semua fitur numerik

```python
# Memilih fitur-fitur numerik
numeric_features = ['carat', 'depth', 'table', 'price', 'x', 'y', 'z']

# Membuat scatter plot untuk semua fitur numerik
sns.pairplot(df[numeric_features])
plt.show()
```

# Heat Map

Prompt ChatGPT : beri kode untuk menampilkan heat map dari fitur-fitur numerik

```python
# Memilih fitur-fitur numerik
numeric_features = ['carat', 'depth', 'table', 'price', 'x', 'y', 'z']

# Menghitung korelasi antar fitur numerik
correlation_matrix = df[numeric_features].corr()

# Membuat heatmap untuk menampilkan korelasi
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True)
plt.title('Heatmap of Numeric Features')
plt.show()
```

# Violin Plot

Prompt ChatGPT : beri kode untuk menampilkan violin plot dari fitur-fitur numerik

```python
# Memilih fitur-fitur numerik
numeric_features = ['carat', 'depth', 'table', 'price', 'x', 'y', 'z']

# Membuat violin plot untuk setiap fitur numerik
plt.figure(figsize=(12, 8))
for i, feature in enumerate(numeric_features, 1):
    plt.subplot(3, 3, i)
    sns.violinplot(x=df[feature])
    plt.xlabel(feature)

plt.tight_layout()
plt.show()
```

# Bar Plot

Prompt ChatGPT : beri kode untuk menampilkan bar plot dari fitur-fitur numerik

```python
# Menentukan jumlah interval dan menghitung interval lebar
num_bins = 10
bin_width = (df['price'].max() - df['price'].min()) / num_bins

# Membuat interval
bins = [df['price'].min() + i * bin_width for i in range(num_bins+1)]

# Mengelompokkan nilai price ke dalam interval
df['price_bin'] = pd.cut(df['price'], bins=bins, include_lowest=True)

# Menghitung frekuensi setiap interval
bar_data = df['price_bin'].value_counts().sort_index()

# Membuat bar plot
plt.figure(figsize=(12, 6))
bar_data.plot(kind='bar')
plt.xlabel('Price Intervals')
plt.ylabel('Frequency')
plt.title('Bar Plot of Price Intervals')
plt.show()
```
